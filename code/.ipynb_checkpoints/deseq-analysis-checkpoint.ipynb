### PCA Analysis

Let's investigate the structure of our data using principal component analysis:

```python
labels = pd.Series(metadata_chp_control, index=counts_chp_control.index)
```

```python
labels.name = 'label'
```

```python
labels.head()
```

```python
counts_chp_control.shape
```

```python
chp_control_dataset = pd.concat([counts_chp_control, labels], axis=1)
```

```python
chp_control_dataset.head()
```

```python
chp_control_dataset['label'].replace(1, 'CHP',inplace=True)
chp_control_dataset['label'].replace(0, 'Control',inplace=True)
```

```python
chp_control_dataset.head()
```

```python
chp_control_dataset.tail()
```

```python
samples = counts_chp_control.index
```

```python
features = counts_chp_control.columns
```

We normalize our data to avoid bias when computing our principal components:

```python
from sklearn.preprocessing import StandardScaler
x = chp_control_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
```

```python
x.shape
```

Given there are 185 samples, we can have at most 185 principal components to explain the variation in our dataset.

```python
np.mean(x),np.std(x)
```

```python
normalized_chp_control = pd.DataFrame(x, columns=features, index=samples)
```

```python
normalized_chp_control
```

```python
from sklearn.decomposition import PCA
pca_chp = PCA(n_components=10)
principalComponents_chp = pca_chp.fit_transform(x)
```

```python
principal_chp_df = pd.DataFrame(data=principalComponents_chp
             , columns = [f'PC{i+1}' for i in range(10)], index=samples)
```

```python
principal_chp_df
```

```python
print('Explained variability per principal component: {}'.format(pca_chp.explained_variance_ratio_))
```

Let's create a Scree plot to visualize the amount of variation explained in our dataset for each PC:

```python
plt.figure(figsize=(10,8))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.ylabel('fraction of variation explained',fontsize=16)
categories = [f'PC{i+1}' for i in range(10)]
values = pca_chp.explained_variance_ratio_
plt.bar(categories, values)
plt.title("Scree Plot for CHP Dataset",fontsize=20)

if SAVE:
    plt.savefig(OUTPUT_PATH + 'scree_plot_for_chp_dataset.png')
```

```python
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('PC1',fontsize=20)
plt.ylabel('PC2',fontsize=20)
plt.title("PCA for CHP Dataset",fontsize=20)
targets = ['CHP', 'Control']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = chp_control_dataset['label'] == target
    # print(indicesToKeep)
    plt.scatter(principal_chp_df.loc[indicesToKeep, 'PC1']
               , principal_chp_df.loc[indicesToKeep, 'PC2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})

if SAVE:
    plt.savefig(OUTPUT_PATH + 'pca_results_for_chp_dataset.png')
```

The scatterplot above shows the CHP and control samples are linearly separable up to some extent. While CHP samples tend to cluster towards the left and control samples tend to cluster towards the right, some samples from both classes overlap, indicating that the first two principal components are not sufficient to distinguish between the two classes. That being said, it looks like our first principal component captures the biological signal in our dataset that we would like to test, given the separation between CHP and control samples. 

### Differential Expression Analysis

Let's start by doing a single factor analysis of our data:

```python
counts_df = counts_chp_control
```

```python
counts_df.shape
```

```python
counts_df.index
```

```python
counts_df.head()
```

```python
labels_df = pd.DataFrame(labels).rename(columns={'label': 'condition'})
```

```python
labels_df.shape
```

```python
labels_df.head()
```

Remove samples that are missing an annotation for condition:

```python
samples_to_keep = ~labels_df.condition.isna()
counts_df = counts_df.loc[samples_to_keep]
labels_df = labels_df.loc[samples_to_keep]
```

```python
labels_df.shape
```

Remove genes that have less than ten read counts in total:

```python
genes_to_keep = counts_df.columns[counts_df.sum(axis=0) >= 10]
counts_df = counts_df[genes_to_keep]
```

```python
counts_df.shape
```

Next, we will compare gene expression of samples from the CHP class to samples from the control class using differential expression analysis. In this case, we are only using the condition column as a design factor in our model:

```python
inference = DefaultInference(n_cpus=8)
dds = DeseqDataSet(
    counts=counts_df,
    metadata=labels_df,
    # design='~condition', # this is supported from v0.5.0pre1 onwards
    refit_cooks=True,
    inference=inference,
    # n_cpus=8, # n_cpus can be specified here or in the inference object
)
```

```python
dds.deseq2()

if SAVE:
    with open(os.path.join(OUTPUT_PATH, "dds.pkl"), "wb") as f:
        pkl.dump(dds, f)
```

```python
print(dds)
```

```python
print(dds.varm['dispersions'])
```

```python
print(dds.varm['LFC'])
```

```python
labels_df
```

```python
ds = DeseqStats(dds, contrast=["condition", "1", "0"], inference=inference)
```

```python
ds.summary()

if SAVE:
    with open(os.path.join(OUTPUT_PATH, "ds.pkl"), "wb") as f:
        pkl.dump(ds, f)
```

Let's visualize our results using a volcano plot:

```python
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
import plotly.express as px

def interactive_volcano(
    res_df,
    lfc_col="log2FoldChange",
    padj_col="padj",
    gene_col=None,          # if None, use index
    lfc_thresh=1.0,
    padj_thresh=0.05,
    top_n=10,
    title=""
):
    df = res_df.copy()

    # Gene names
    if gene_col is None:
        df["gene"] = df.index
    else:
        df["gene"] = df[gene_col]

    # Remove missing
    df = df.dropna(subset=[lfc_col, padj_col])

    # Compute -log10(padj)
    df["neglog10_padj"] = -np.log10(df[padj_col])

    # Significance category
    df["category"] = "Not significant"
    df.loc[(df[padj_col] < padj_thresh) & (df[lfc_col] > lfc_thresh), "category"] = "Upregulated"
    df.loc[(df[padj_col] < padj_thresh) & (df[lfc_col] < -lfc_thresh), "category"] = "Downregulated"

    # Rank genes by significance for labeling
    df["rank_metric"] = df["neglog10_padj"] * np.abs(df[lfc_col])
    top_genes = df.nlargest(top_n, "rank_metric")

    # Plot
    fig = px.scatter(
        df,
        x=lfc_col,
        y="neglog10_padj",
        color="category",
        hover_data=["gene", lfc_col, padj_col],
        title=title
    )

    # Label top genes
    fig.add_scatter(
        x=top_genes[lfc_col],
        y=top_genes["neglog10_padj"],
        mode="text",
        text=top_genes["gene"],
        textposition="top center",
        showlegend=False
    )

    # Threshold lines
    fig.add_vline(x=lfc_thresh, line_dash="dash")
    fig.add_vline(x=-lfc_thresh, line_dash="dash")
    fig.add_hline(y=-np.log10(padj_thresh), line_dash="dash")

    fig.update_layout(
        xaxis_title="log2 Fold Change",
        yaxis_title="-log10 adjusted p-value"
    )

    fig.show()
    return fig
```

```python
fig = interactive_volcano(ds.results_df)
```
